## Criando Features

```python
#Importações necessárias
import pandas as pd
import numpy as np
```

```python
#Carregando o dataset
df = pd.read_csv('../data/raw/ecommerce_data.csv')
```

```python
#Convertendo a coluna 'OrderDate' para o formato datetime
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
```

```python
#Criando colunas temporais 
df['Year'] = df['OrderDate'].dt.year #Extraindo o ano da data de pedido
df['Month'] = df['OrderDate'].dt.month #Extraindo o número do mês da data de pedido
df['MonthName'] = df['OrderDate'].dt.month_name() #Extraindo o nome do mês da data de pedido
df['WeekDay'] = df['OrderDate'].dt.day_name() #Extraindo o nome do dia da semana da data de pedido
```

```python
#Criando métricas auxiliares

#Coluna com o valor médio do ticket
df['AverageTicket'] = df['TotalAmount'] / df['Quantity']
```

```python
#Coluna categórica de faixa de preço
df['PriceRange'] = pd.qcut(
    df['Price'],
    q=3,
    labels=['Low', 'Medium', 'High']
)
```

```python
#Coluna categórica de faturamento
df['RevenueLevel'] = pd.qcut(
    df['TotalAmount'],
    q=3,
    labels=['Low', 'Medium', 'High']
)
```

```python
#Criando colunas de classificação de avaliação
df['RatingCategory'] = pd.cut(
    df['Rating'],
    bins=[0, 2, 4, 5],
    labels=['Poor', 'Good', 'Excellent']
)
```

```python
#Validando a consistência dos dados

#Coluna de faturamento calculada
df['CalculatedTotal'] = df['Price'] * df['Quantity']
```

```python
#Validando a consistência entre o faturamento calculado e o faturamento registrado

consistency_check = (
    df['CalculatedTotal'].round(2)
    ==
    df['TotalAmount'].round(2)
)

consistency_check.value_counts()
```
