## Business Insights

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carregando os dados processados
df = pd.read_csv('../data/processed/ecommerce_processed.csv')
```

```python
#Convertendo a coluna de data para datetime
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
```

```python
#Insights de importância para o negócio

#Receita por categoria
category_revenue = (
    df.groupby('Category')['TotalAmount']
    .sum()
    .sort_values(ascending=False)
)

category_revenue
```

```python
#Visualização: Receita por categoria
category_revenue = (
    df.groupby('Category')['TotalAmount']
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(10,5))

category_revenue.plot(kind='bar')

plt.title('Revenue by Category')
plt.ylabel('Revenue')

plt.savefig(
    '../reports/figures/Business_Insights/revenue_by_category.png',
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

## Conclusão da Etapa de Business Intelligence (BI)

Com base nas análises de dados realizadas, levantamos os seguintes insights estratégicos e oportunidades de ação para o negócio:

* **Categorias e Produtos Curva A:** A receita é fortemente impulsionada pelas categorias de **Electronics** (R$ 92.3M) e **Fashion** (R$ 91.5M). O produto *Xiaomi Redmi Note 12* é o carro-chefe isolado. 
    * *Ação:* Direcionar maior verba de marketing para essas categorias, garantir controle rígido de estoque do produto campeão e criar campanhas de *cross-sell* com categorias menores (Wearables e Accessories).
* **Segmentação Premium:** Produtos com alto ticket médio (ex: Xiaomi Redmi Note e tênis esportivos acima de R$ 10k) representam uma fatia altamente rentável.
    * *Ação:* Criar estratégias de atendimento VIP, jornadas de compras personalizadas e vantagens logísticas (frete grátis/expresso) para aumentar a conversão.
* **Geografia e Canais de Venda:** O mercado é muito equilibrado territorialmente (com leve vantagem para Cairo e Casablanca) e digitalmente (Souq, Jumia e Amazon com receitas na faixa de R$ 100M).
    * *Ação:* Manter a diversificação omnichannel, mas analisar qual plataforma oferece as melhores margens (descontadas as taxas) para priorizar anúncios nela. Otimizar a logística no norte da África (Egito e Marrocos).
* **Experiência do Cliente:** As notas médias orbitam na faixa de 3.0 para quase todos os produtos, e não há correlação entre volume de avaliações e receita.
    * *Ação:* Focar em descobrir a raiz da "experiência mediana" (entrega, qualidade ou atendimento) e trabalhar na melhoria do serviço, visto que produtos vendem pela demanda, mas a fidelização exigirá notas maiores.
* **Sazonalidade e Outliers (B2B):** As vendas são estáveis ao longo do ano, porém Dezembro foi o mês mais fraco (possível ruptura de estoque no fim de ano). Além disso, há registros de compras na casa de R$ 100 mil.
    * *Ação:* Avaliar o que causou a queda em Dezembro para corrigir a rota no próximo ano. Investigar clientes que fazem pedidos gigantes, pois indicam uma grande oportunidade de abrir um braço de vendas B2B (Atacado/Corporativo).
