# Entendendo os dados 

```python
#Importações necessárias
import pandas as pd
import numpy as np
```

```python
#Carregando o dataset
df = pd.read_csv('../data/raw/ecommerce_data.csv')
```

```python
#Analisando as primeiras linhas do dataset
df.head(10)
```

```python
#Analisando as últimas linhas do dataset
df.tail(10)
```

```python
#Vendo as dimensões do dataset
print(f"O dataset possui {df.shape[0]} linhas e {df.shape[1]} colunas.")
```

```python
#Analisando os tipos das colunas
print("Tipos das colunas:")
print(df.dtypes)
```

```python
#Informações gerais do dataset
df.info()
```

```python
#Analisando as estatísticas descritivas do dataset
df.describe()
```

```python
#Verificando nulos no dataset
print("Número de valores nulos por coluna:")
print(df.isnull().sum())    
```

```python
#Verificando valores duplicados no dataset
print(f"Número de linhas duplicadas: {df.duplicated().sum()}")  
```

```python
#Verificando a quantidade de valores únicos em cada coluna
print("Quantidade de valores únicos em cada coluna:")
print(df.nunique())
```

```python
#Verificando a frequência de cada categoria na coluna 'Category'
print("Frequência de cada categoria na coluna 'Category':")
print(df['Category'].value_counts())
```

```python
#Verificando a distribuição de valores
df['Price'].describe()
```

```python
#Conferindo o formato da coluna 'OrderDate'
df['OrderDate'].head()
```

```python
#Convertendo a coluna 'OrderDate' para o tipo datetime
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
```

```python
#Conferindo se a conversão foi bem-sucedida
df['OrderDate'].head()
```

```python
#Conferindo o intervalo de datas
print(f"Data mínima: {df['OrderDate'].min()}")
print(f"Data máxima: {df['OrderDate'].max()}")
```

```python
#Exemplo aleátorio de uma linha do dataset
print("Exemplo aleatório de uma linha do dataset:")
print(df.sample(3))
```
