## 📊 EDA — Exploratory Data Analysis

```python
#Importações necessárias 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/processed/ecommerce_processed.csv')
```

```python
#Definindo o tipo de dado da coluna 'OrderDate' como datetime novamente
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
```

```python
#KPIs

#Receita total
df['TotalAmount'].sum()
```

```python
#Quantidade total de vendida
df['Quantity'].sum()
```

```python
#Ticket médio geral
df['AverageTicket'].mean()
```

```python
#Número de pedidos únicos
df['OrderID'].nunique()
```

```python
#Análises de produtos

#Produto mais vendido
df.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
```

```python
#Produtos com maior receita
df.groupby('Product')['TotalAmount'].sum().sort_values(ascending=False) 
```

```python
#Produtos com as melhores avaliações
df.groupby('Product')['Rating'].mean().sort_values(ascending=False)
```

```python
#Analisando por categorias

#Receita por categoria
df.groupby('Category')['TotalAmount'].sum().sort_values(ascending=False)
```

```python
#Quantidade vendida por categoria
df.groupby('Category')['Quantity'].sum().sort_values(ascending=False)
```

```python
#Média de ticket por categoria
df.groupby('Category')['AverageTicket'].mean()
```

```python
#Análise por plataforma

#Receita por plataforma
df.groupby('Platform')['TotalAmount'].sum()
```

```python
#Média de ranting por plataforma
df.groupby('Platform')['Rating'].mean()
```

```python
#Analisando por cidade

#Ranking por cidades
df.groupby('City')['TotalAmount'].sum().sort_values(ascending=False)
```

```python
#Pedidos por cidade
df['City'].value_counts()
```

```python
#Analise por avaliações

#Média de ranting
df['Rating'].mean()
```

```python
#Distribuição de ratings
df['RatingCategory'].value_counts()
```

```python
#Relação entre avaliação e receita
df[['Reviews', 'TotalAmount']].corr()
```

```python
#Análises temporais

#Receita por mês
df.groupby('MonthName')['TotalAmount'].sum()
```

```python
#Quantidade vendida por mês
df.groupby('MonthName')['Quantity'].sum()
```

```python
#Receita por dia da semana
df.groupby('WeekDay')['TotalAmount'].sum()
```

```python
#Correlações 

#Correlação numérica
df.corr(numeric_only=True)
```

```python
#Heatmap
plt.figure(figsize=(10,6))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')

# Salvar figura
plt.savefig(
    '../reports/figures/EDA/correlation_heatmap.png',
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

```python
#Visualizações

#Receita por categoria
plt.figure(figsize=(10,5))

sns.barplot(
    data=df,
    x='Category',
    y='TotalAmount',
    estimator=sum
)

plt.title('Revenue by Category')

plt.savefig(
    '../reports/figures/EDA/revenue_by_category.png',
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

```python
#Top produtos
top_products = (
    df.groupby('Product')['TotalAmount']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,5))

top_products.plot(kind='bar')

plt.title('Top Products by Revenue')

plt.savefig(
    '../reports/figures/EDA/top_products_revenue.png',
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

```python
#Distribuição de preços
plt.figure(figsize=(10,5))

sns.histplot(
    df['Price'],
    bins=30
)

plt.title('Price Distribution')

plt.savefig(
    '../reports/figures/EDA/price_distribution.png',
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

```python
#Evolução mensal da receita
monthly_sales = (
    df.groupby('Month')['TotalAmount']
    .sum()
)

plt.figure(figsize=(10,5))

monthly_sales.plot()

plt.title('Monthly Revenue')

plt.savefig(
    '../reports/figures/EDA/monthly_revenue.png',
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

```python
#Receita por cidade 
city_revenue = (
    df.groupby('City')['TotalAmount']
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(10,5))

city_revenue.plot(kind='bar')

plt.title('Revenue by City')

plt.savefig(
    '../reports/figures/EDA/revenue_by_city.png',
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

```python
#Receita por plataforma
platform_revenue = (
    df.groupby('Platform')['TotalAmount']
    .sum()
)

plt.figure(figsize=(8,5))

platform_revenue.plot(kind='bar')

plt.title('Revenue by Platform')

plt.savefig(
    '../reports/figures/EDA/revenue_by_platform.png',
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

```python
#Distribuição por avaliações
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='RatingCategory'
)

plt.title('Rating Distribution')

plt.savefig(
    '../reports/figures/EDA/rating_distribution.png',
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

# Conclusões da Análise Exploratória (EDA)

- O dataset apresentou mais de 301 milhões em receita total distribuídos em 10.000 pedidos.
- As categorias `Electronics` e `Fashion` lideraram tanto em faturamento quanto em quantidade vendida.
- O produto `Xiaomi Redmi Note 12` apresentou o maior faturamento e volume de vendas.
- As plataformas apresentaram faturamentos bastante equilibrados, com leve destaque para a `Souq`.
- As cidades `Cairo` e `Casablanca` concentraram a maior receita do dataset.
- As análises temporais indicaram estabilidade nas vendas ao longo do ano, sem sazonalidade extrema.
- As correlações mostraram forte relação entre `Price`, `Quantity` e `TotalAmount`.
- Não foram observadas relações relevantes entre avaliações/reviews e faturamento.
- Foram identificados possíveis outliers financeiros em `Price` e `TotalAmount`.
- O dataset apresentou comportamento consistente e grande potencial para geração de insights de negócio.
